# Depth Analysis

```python
Col = ['tab:red', 'tab:green', 'tab:blue', 'tab:purple', 'tab:pink', 'tab:cyan']

L1 = [1,3,6,7,8,9,10,18,19,23,27,29] # Retro
#L2 = [21] # Slope Sea avant retro
L3 = [2,17,21,22,25] # Slope Sea
L4 = [26] # Labrador Sea
L5 = [0,4,5,12,13,14,15,16,20] # Labrador Shelf
L6 = [11,28] # South
L7 = [24] # Belle Isle
Ls = [L1,L3,L4,L5,L6,L7]
labels = ['Retroflected', 'Westward-flowing', 'Labrador Sea', 'Labrador Shelf', 'Southward-flowing', 'Belle Isle']
```

```python
def longitude_profondeur(labels, lats, lons, depths, path_save, Ls, names):      
    s = lats.shape
    
    # depths in Glorys
    dg = [0,1,2,3,5,6,7,9,11,13,15,18,21,25,29,34,40,47,55,65,77,92,109,130,155,186,222,266,318,380,453,541,643,763,902,1062,1245,1452,1684,1941,2225,2533,2865,3220,3597,3992,4405,4833,5274,5727]
    bins = dg[22:]  

    # --- check the depths in every cluster --- #

    dall_all = [] ; itraj_all = [] ; c=0; dall_for_perc_all = []
    dall = [] ; itraj = [] ; dall_formean = []; d_all_perc = []
    for superclust in Ls:
        for n in range(len(labels)):
            if labels[n] in superclust :
                d = depths[n,:][~np.isnan(depths[n,:])]
                dall_formean.append(depths[n,:])
                dall.extend(d)
                d_all_perc.append(d)
                itraj.extend(range(len(d)))

    dall_all.append(dall)
    itraj_all.append(itraj)
    dall_for_perc_all.append(d_all_perc)
    c+=1
    
    # Figure looking at longitude
    f = plt.figure(figsize = (4,4))
    ax = f.add_subplot(1,1,1)
    count = 0
    for superclust in Ls:
        ALL_LONS = []
        ALL_DEPTHS = []
        for n in range(len(labels)):
            if labels[n] in superclust :
                ALL_LONS.append(lons[n,:])
                ALL_DEPTHS.append(depths[n,:])
        ALL_LONS = np.array(ALL_LONS)
        ALL_DEPTHS = np.array(ALL_DEPTHS)
        
        ax.plot(np.nanmean(ALL_LONS, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), c=Col[count], label=names[count])  
        #ax.plot(np.nanpercentile(ALL_LONS, 10, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), ':', c=Col[count], alpha=0.5)
        if len(ALL_DEPTHS) > 1 :
            ax.fill_betweenx(np.nanmean(ALL_DEPTHS, axis = 0), np.nanpercentile(ALL_LONS, 10, axis = 0), np.nanpercentile(ALL_LONS, 90, axis = 0), facecolor=Col[count], alpha=0.1)
        #ax.plot(np.nanpercentile(ALL_LONS, 90, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), ':', c=Col[count], alpha=0.5)     
        
        count+=1
        
    ax.set_ylabel('Depth [m]')
    ax.set_xlabel('Longitude')
    ax.yaxis.set_major_locator(plt.MaxNLocator(4))
    plt.gca().invert_yaxis()
    plt.tight_layout()
    #plt.legend()
    plt.savefig(path_save+'depths_lons_all.png',dpi=300)
    plt.show()
    
    # Figure looking at depth
    f = plt.figure(figsize = (4,7))
    ax = f.add_subplot(1,1,1)
    count = 0
    for superclust in Ls:
        ALL_LONS = []
        ALL_DEPTHS = []
        for n in range(len(labels)):
            if labels[n] in superclust :
                ALL_LONS.append(lons[n,:])
                ALL_DEPTHS.append(depths[n,:])
        ALL_LONS = np.array(ALL_LONS)
        ALL_DEPTHS = np.array(ALL_DEPTHS)
        
        lw = 0.5
        alpha = 0.3
        lt = '-' # line type
        
        ax.plot(np.nanmean(ALL_LONS, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), c=Col[count], label=names[count])  
        ax.plot(np.nanmean(ALL_LONS, axis = 0), np.nanpercentile(ALL_DEPTHS, 10, axis = 0), lt, c=Col[count], alpha=alpha, lw=lw)
        ax.plot(np.nanpercentile(ALL_LONS, 10, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), lt, c=Col[count], alpha=alpha, lw=lw)
        #if len(ALL_DEPTHS) > 1 :
        #    ax.fill_between(np.nanmean(ALL_LONS, axis = 0), np.nanpercentile(ALL_DEPTHS, 10, axis = 0), np.nanpercentile(ALL_DEPTHS, 90, axis = 0), facecolor=Col[count], alpha=0.1)
        ax.plot(np.nanmean(ALL_LONS, axis = 0), np.nanpercentile(ALL_DEPTHS, 90, axis = 0), lt, c=Col[count], alpha=alpha, lw=lw)
        ax.plot(np.nanpercentile(ALL_LONS, 90, axis = 0), np.nanmean(ALL_DEPTHS, axis = 0), lt, c=Col[count], alpha=alpha, lw=lw)

        count+=1
        
    ax.set_ylabel('mean depth')
    ax.set_xlabel('mean longitude')
    #plt.legend()
    plt.savefig(path_save+'depths_lons_2_all.png',dpi=300)
    plt.show()
```

```python
# Depth with distance along trajectory 

mean_med = np.nanmean
perc = True
names = labels
labels = labels_all
lats = lats_all
lons = lons_all
dists = dists_along
depths = depths_all

do = True
if do :
    listlabels = list(set(labels))
    s = lats.shape

    # --- average the depth --- #
    dd_mean_all = [] ; dd_10_all = [] ; dd_90_all = [] ; sizes=[]
    ii=0
    for superclust in Ls:
        ii+=1
        
        d_dist = 500.
        distlist = np.arange(0,16000,d_dist)
        depth_dist = [[] for i in distlist] # initialize list of list for distances bins
        for n in range(len(labels)):
            if labels[n] in superclust :
                
                #d = depths[n,:][~np.isnan(depths[n,:])]
                d = depths[n,:]
                #di = dists[n,:][~np.isnan(dists[n,:])]
                di = dists[n,:]
                
                for i in range(len(d)):
                    if np.isnan(di[i]) == False:
                        idx = int(di[i] / d_dist)
                        try:
                            depth_dist[idx-1].append( d[i] )
                        except:
                            print(idx, len(depth_dist))
                    
        # calculate mean and other stats
        dd_mean = [] ; dd_10 = [] ; dd_90 = [] ; size = []
        for i in range(len(depth_dist)):
            dd_mean.append( np.mean(depth_dist[i]) )
            dd_10.append( np.nanpercentile(depth_dist[i], 10) )
            dd_90.append( np.nanpercentile(depth_dist[i], 90) )
            size.append( len(depth_dist[i])/10000. )
            
        dd_mean_all.append(dd_mean)
        dd_10_all.append(dd_10) ; dd_90_all.append(dd_90)
        sizes.append(size)
    
    # save
    np.savetxt(path_save+'distlist.csv', distlist, delimiter=',')
    np.savetxt(path_save+'depth_mean.csv', dd_mean_all, delimiter=',')
    np.savetxt(path_save+'depth_10perc.csv', dd_10_all, delimiter=',')
    np.savetxt(path_save+'depth_90perc.csv', dd_90_all, delimiter=',')
    
    # --- plot depending on their classif --- #
    
    f = plt.figure(figsize = (8,3)) 
    ax = f.add_subplot(1,1,1)
    
    for i in range(len(Ls)):
        ax.fill_between(distlist, dd_10_all[i], dd_90_all[i], facecolor=Col[i], alpha=0.2)    
    for i in range(len(Ls)):
        ax.plot(distlist, dd_mean_all[i], c= Col[i], label=names[i])
        sp = ax.scatter(distlist, dd_mean_all[i], s=sizes[i], edgecolors=Col[i], facecolors='none')
        sp = ax.scatter(distlist, dd_mean_all[i], s=sizes[i], edgecolors=Col[i], facecolors='none')
        ax.scatter(distlist, dd_mean_all[i], s=5, c=Col[i])
        
    plt.legend(loc=(1.04,0.1))
    plt.legend(*sp.legend_elements("sizes", num=3), title=r'$\times$10 000 particles')
    ax.set_ylabel('Depth (m)')
    ax.set_xlabel('Distance (km)')
    ax.invert_yaxis()
    ax.set_ylim([2100,0]) #; axns[1].set_ylim([5000,0])
    ax.set_xlim([-100,10000])

    plt.tight_layout()
    plt.savefig(path_save+'depths_along_all.png', dpi=300)
    plt.show()
    plt.close()
```

```python
# density plot
import matplotlib.colors as colors
import seawater.eos80 as sw

# Histogram
x = np.arange(30,43,1./12.)
y = np.arange(0,2000,10.)
    
hist, xe, ye = np.histogram2d(ALL_LATS, ALL_DEPTHS, bins=(x,y))
hist[hist==0] = np.nan

# load Glorys 
ds = xr.open_dataset('/storage3/shared/Glorys12/withw/GLOBAL_REANALYSIS_PHY_001_030_2018_12_01_v2_w.nc')
ds = ds.sel(latitude=slice(30,43), depth=slice(0,2000))
ds = ds.sel(longitude=-50, method='nearest')

temp = ds.thetao.values
sal = ds.so.values
pden = np.zeros(temp.shape)
for i in range(temp.shape[0]):
    for j in range(temp.shape[1]):
        pden[i,j] = sw.pden(sal[i,j], temp[i,j], sw.pres(ds.depth[i], ds.latitude[j]))-1000.
```
